```python
import pandas as pd
#importing the dataset
to_melt = pd.read_excel('Baby_Care_source.xlsx', sheet_name='to_melt')
to_melt
```

```python
#melting the columns year, volume, average and values
pass_to_excel = pd.melt(to_melt,id_vars=['Product/item', 'MANUFACTURER', 'BRAND', 'PACKAGE TYPE',
       'PROTECTION / CARE', 'PACKAGE SIZE'],value_vars=['Val 2022', 'Val 2023',
       'Val YTD 24', 'Vol 2022', 'Vol 2023', 'Vol YTD 24', 'AVR 2022',
       'AVR 2023', 'AVR YTD 24'],var_name='Val_Year', value_name='Value').sort_values(['Product/item','Val_Year']).reset_index().drop('index', axis=1)
pass_to_excel.head(10)
```

```python
pass_to_excel.to_excel('Baby_Care_project.xlsx', sheet_name='Dataset', index=False)
```
